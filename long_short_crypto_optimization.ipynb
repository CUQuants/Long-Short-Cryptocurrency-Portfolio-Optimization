```python
fig, axes = plt.subplots(3,3, figsize = (20,20))

#the index to reference the dataframe
k = 0

#loop through all subplots
for i in range(3):
    for j in range(3):
        
        #plot returns
        returns[returns.columns[k]].plot(kind = "hist", bins = 100, ax = axes[i,j], grid = True, title = returns.columns[k])
        k += 1
```

```python
ef = Efficient_Frontier(returns, returns.columns)
```

```python
num_portfolios = 100000
rf = 0.0
results_frame = ef.simulate_random_portfolios("mean_returns", "covariance", num_portfolios, rf)
```

```python
results_frame
```

```python
portfolios = ef.find_portfolios(results_frame)
portfolios
```

```python
plt.figure(figsize = (17,9))
plt.scatter(results_frame['stdev'], results_frame['ret'], c = results_frame.sharpe, cmap = "RdYlBu")
plt.scatter(portfolios[0][1], portfolios[0][0], marker = (5,1,0), color = 'r', s = 500)
plt.scatter(portfolios[1][1], portfolios[1][0], marker = (5,1,0), color = 'g', s = 500)

plt.title("Efficient Frontier")
plt.xlabel("Standard Deviation")
plt.ylabel("Expected Returns")
plt.grid()


plt.show()
```
